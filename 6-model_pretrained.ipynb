模型的保存和加载

```python
import torch
vgg16 = torchvision.models.vgg16(pretrained = False)

torch.save(vgg16, "vgg16_method1.pth")      ## 保存了结构和参数
model = torch.load("vgg16_method1.pth")
print(model)

torch.save(vgg16.state_dict(), "vgg16_method2.pth")    ## 把参数保存成字典，只保存参数，官方推荐
vgg16 = torchvision.models.vgg16(pretrained = False)
vgg16.load_state_dict(torch.load("vgg16_method2.pth"))  ## 添加参数
print(model)
```

```python
# 陷阱1
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()   ## 调用父类的构造函数，否则会默认子类重写了构造函数
        self.conv1 = nn.Conv2d(1, 20, 5)

    def forward(self, x):               ## 在每一个子类中都要重写forward
        x = self.conv1(x)
        return x

model = Model()
torch.save(model, "model.pth")
# 自定义模型时要load的文件中必须包含model的完整结构
model = torch.load("model.pth")
print(model)
```
